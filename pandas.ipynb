```python
#Matplotlib
data={
    "Salary" :[10000,12500,50000,30000,45000,54000,67000,70000,55000,30000,25000,45000,90000,80000,65000]
}
df2=pd.DataFrame(data)
df2
```

```python
#line plot
plt.plot(df2['Salary'],color="black",marker="o",linestyle=":",linewidth="2")
plt.grid()
```

```python
# Histogram
plt.hist(df2['Salary'],bins=7,color="red")
plt.show()
```

```python
#Boxplot
plt.boxplot(df2['Salary'])
plt.show()
```

```python
df2.drop(index=60,inplace=True)
df2
```

```python
#Univariate: Categorical
```

```python
df2["dept"] = ['HR','IT','Finance','HR','IT','Finance','HR','Finance','HR','IT','Finance','Finance','HR','IT','Finance']
```

```python
df2.shape
```

```python
#pichat
count=df2['dept'].value_counts()
count
```

```python
plt.pie(count,labels=count.index,autopct="%1.2f",explode=[0,0.1,0.2])
plt.axis("equal")
plt.show()
```

```python
#Countplot
plt.bar(count.index,count,color=["red","green","yellow"])
plt.show()
```

```python
sort_age =df2.sort_values("Age")
```

```python
#Scatter plot:

plt.scatter(df2["Age"],df2["Salary"],color ="red",marker="*")
plt.grid()
plt.show()
```

```python
#line plot:
plt.plot(sort_age["Age"],df2["Salary"],color ="red",marker="*",linewidth="2")
plt.grid()
plt.show()
```

```python
#bar chart:
plt.bar(sort_age["Age"],df2["Salary"],color = "pink",align="center")
plt.show()
```
